```python
import os

os.environ["XLA_PYTHON_CLIENT_MEM_FRACTION"]="1.00"
os.environ['KERAS_BACKEND'] = 'jax'
import keras
import keras_hub
import numpy as np
import matplotlib.pyplot as plt
```

```python
text_to_image = keras_hub.models.StableDiffusion3TextToImage.from_preset(
    "stable_diffusion_3_medium", width=512, height=512, dtype="bfloat16"
)
```

```python
image = text_to_image.generate(
    "Astronaut on a tree, cold color palette, muted colors, detailed, 8k"
)
plt.imshow(image)
```
